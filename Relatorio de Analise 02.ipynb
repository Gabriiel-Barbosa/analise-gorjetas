Pedir sobremesa altera o valor da gorjeta ? 

```python
categorico_sobremesas = sns.catplot(x='sobremesa', y='gorjeta', data=gorjetas)
```

O gráfico categórico não retornou nenhum dado relevante sobre a infliuência da sobremesa na gorjeta

```python
dispercao_comparativo = sns.relplot(x='valor_da_conta', y='gorjeta', hue='sobremesa', col='sobremesa', data=gorjetas)
```

```python
analise_grafica1 = sns.lmplot(x='valor_da_conta', y='gorjeta', col='sobremesa', hue='sobremesa', data=gorjetas)
```

```python
analise_grafica2 = sns.lmplot(x='valor_da_conta', y='porcentagem', col='sobremesa', hue='sobremesa', data=gorjetas)
```
